```python
import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize=(5,5)) #生成畫板

x = np.linspace(0,1,100)
y = x**2
z = x

ax = plt.subplot(2,2,1) #在畫板哪個位置
plt.plot(x,y)

ax = plt.subplot(2,2,2)
plt.plot(x,z)

ax = plt.subplot(2,2,3)
plt.plot(x,y)
plt.plot(x,z)
plt.show #顯示在同張圖上
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1,1,100)
plt.figure(figsize=(2,2))

for y in range(0,4): #學以致用的for～
  plt.plot(x,x**y)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize=(2,2))

pi = np.pi
x = np.linspace(-2*pi,2*pi,100)
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize=(3,3))

x = np.linspace(-4,4,100)
for a in [1,1.5,2]:
  plt.plot(x,np.exp(-x**2/a**2),label=f'a={a}') #最後是指標
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-4,4,100)
plt.figure(figsize=(2,2))

plt.plot(x,np.exp(x))
plt.plot(x,np.log(x))
plt.plot([4,-4],[4,-4])

plt.xlim(-4,4)
plt.ylim(-4,4)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
pi = np.pi
plt.figure(figsize=(2,2))

r = 1
theta = np.linspace(0,2*pi,300)
x = r*np.cos(theta)
y = r*np.sin(theta)

plt.plot(x,y)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

r = 1 #半徑大小
w = 2 #總圈數
vz = 2 #高度
t = np.linspace(0,10,300) #每個點

x = r*np.cos(w*t)
y = r*np.sin(w*t)
z = vz*t

ax = plt.subplot(projection = '3d')
ax.plot(x,y,z)
plt.show()
```
